```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

x=np.array([350,1070,630,890,940,500])
y=np.array([520,1600,1000,850,1350,490])

xmean=np.mean(x)
ymean=np.mean(y)

xi=x-xmean
yi=y-ymean

print(xi)
print(yi)

a1=np.sum(xi*yi)/(np.sum(xi*xi))
print(a1)
a0=ymean-a1*xmean
print(a0)
ux=790
uy=a0+a1*ux
print("if student study for",790,"hr. then he will obtain marks=",uy)
yp=a0+a1*x
plt.scatter(x,y)
plt.plot(x,yp)
plt.plot(ux,uy,marker='*')
plt.title("Study vs Marks ")
plt.xlabel("Study Time")
plt.ylabel("Marks Obtained")
plt.show()
```
